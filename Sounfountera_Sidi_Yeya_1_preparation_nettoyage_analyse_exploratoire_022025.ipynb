```python
#Importation du fichier Disponibilite Alimentaire_2017.csv
dispo_alim_2017 = pd.read_csv('Data/DisponibiliteAlimentaire_2017.csv', sep=',')

#Importation du fichier Population_2000_2018.csv
pop_2000_2018 = pd.read_csv('Data/Population_2000_2018.csv', sep=',')

#Importation du fichier Population_2000_2018.csv
aide_aliment = pd.read_csv('Data/aide_alimentaire.csv', sep=',')

#Importation du fichier Population_2000_2018.csv
pib = pd.read_csv('Data/PIB.csv', sep=',')
```

<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2.1 - Analyse exploratoire du fichier aide alimentaire</h3>
</div>

```python
aide_aliment.head()
```

```python
aide_aliment['Produit'].unique()
```

```python
aide_aliment = aide_aliment[(aide_aliment['Année']==2016)].groupby('Pays bénéficiaire')['Valeur'].sum().reset_index()
aide_aliment.rename(columns={"Pays bénéficiaire":"Zone","Valeur":"Aide_ali"}, inplace=True)
aide_aliment.head()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2.2 - Analyse exploratoire du fichier disponibilité alimentaire</h3>
</div>

```python
dispo_alim_2017.head()
```

```python
dispo_alim_2017['Produit'].unique()
```

```python
df_voilaile = dispo_alim_2017[dispo_alim_2017['Produit']=='Viande de Volailles']
```

```python
df_pivot = df_voilaile.pivot(index=['Zone','Année'],columns=['Élément'],values='Valeur').fillna(0).reset_index()
```

```python
#multiplication de toutes les lignes contenant des milliers de tonnes en Kg
a_convertir = ['Disponibilité intérieure','Exportations - Quantité','Importations - Quantité','Nourriture','Production']
for c in a_convertir:
    df_pivot[c] = df_pivot[c] * 1000000
```

```python
colonnes_a_garder = ['Zone','Disponibilité intérieure','Exportations - Quantité','Importations - Quantité','Nourriture','Production']
df_alim_nett = df_pivot[colonnes_a_garder]
```

```python
df_alim_nett.head()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2.3 - Analyse exploratoire du fichier population</h3>
</div>

```python
pop_2000_2018.head()
```

```python
df_pop_2017_2018 = pop_2000_2018[pop_2000_2018['Année'].isin([2017, 2018])]
```

```python
df_pop_2017_2018['Année'].unique()
```

```python
df_pivot_pop = df_pop_2017_2018.pivot(index=['Zone'],columns=['Année'],values='Valeur').fillna(0).reset_index()
df_pivot_pop['Variation_population']=round((df_pivot_pop[2018] - df_pivot_pop[2017]) / df_pivot_pop[2017] * 100,2)
df_pivot_pop.drop(columns=[2017,2018],axis=1,inplace=True)
df_pivot_pop.head()
```

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">2.4 - Analyse exploratoire du fichier PIB</h3>
</div>

```python
pib.head()
```

```python
df_pib_2017 = pib[pib['Année'] == 2017]
colonnes_a_garder = ['Zone','Valeur']
pib_2017 = df_pib_2017[colonnes_a_garder]
pib_2017.columns=['Zone','PIB']
pib_2017.head()
```

<div style="background-color: RGB(51,165,182);" >
<h2 style="margin: auto; padding: 20px; color:#fff; ">Etape 3 - Jonction des fichiers</h2>
</div>

<div style="border: 1px solid RGB(51,165,182);" >
<h3 style="margin: auto; padding: 20px; color: RGB(51,165,182); ">Etape 3.1 - Jonction du fichier aide_aliment, df_alim_nett, df_pivot_pop, et pib_2017</h3>
</div>

```python
df_final = df_pivot_pop.merge(aide_aliment, on="Zone", how="left")\
               .merge(df_alim_nett, on="Zone", how="left")\
               .merge(pib_2017, on="Zone", how="left")
df_final.fillna(0,inplace=True)
```
